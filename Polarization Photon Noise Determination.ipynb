```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
for i in range(1,10000,10):
    states = np.random.randint(0,6,i)
    y = np.abs(np.mean(states[states<2]))
    plt.semilogy(i, y, '.', c='k')

x = np.arange(1,10000,100)
plt.plot(x, 1/2.*np.ones_like(x))
plt.show()
```

```python
for i in range(1,10000,100):
    states = np.random.randint(0,6,i*1000).reshape(1000,-1).astype(float)
    states[states >= 2] = np.nan
    y = np.std(np.nanmean(states, axis=1))
    plt.semilogy(i, y, '.', c='k')

x = np.arange(1,10000,100)
plt.plot(x, 1/np.sqrt(x/3.)/2.)
plt.show()
```
