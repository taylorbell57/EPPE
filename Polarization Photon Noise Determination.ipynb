```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special
```

```python
def pdf(n,p=0.5):
    i = np.arange(n)
    return scipy.special.comb(n, i)*p**i**(1-p)**(n-i)
```

```python
def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))
```

# Binomial Distribution

```python
n = 1000
x = np.arange(n)
y = pdf(n)
plt.plot(x, y)
std = 15.811388300841896
gaus = gaussian(x, n/2., std)
gaus = gaus*np.max(y)/np.max(gaus)
plt.plot(x, gaus)
plt.show()

plt.plot(x, y-gaus)
plt.show()
```

# Fitting Gaussian to Random Numbers

```python
iters = 100000
n = 1000
states = np.random.randint(0,2,iters*n).reshape(iters,n)
```

```python
plt.hist(np.sum(states, axis=1), density=True, bins=20)

std = np.sqrt(n)/2.
x = np.arange(n/2-3*std, n/2+3*std)
gaus = gaussian(x, n/2., np.sqrt(n)/2.)
gaus = gaus*plt.gca().get_ylim()[1]*0.95/np.max(gaus)
plt.plot(x, gaus)

plt.show()
```

# Fitting Gaussian to 6 Random States

```python
iters = 10000
n = 1000
states = np.random.randint(0,6,iters*n).reshape(iters,n).astype(float)
states[states>1] = np.nan
```

```python
plt.hist(np.nansum(states, axis=1), density=True, bins=20)

std = np.sqrt(n/3)/2.
x = np.arange(n/3/2-3*std, n/3/2+3*std)
gaus = gaussian(x, n/3/2., np.sqrt(n/3)/2.)
gaus = gaus*plt.gca().get_ylim()[1]*0.95/np.max(gaus)
plt.plot(x, gaus)

plt.show()
```
